```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Bài 1 :Dùng Pandas đọc file.

datasource=pd.read_csv('./weight-height.csv')
datasource.head()
```

```python
# Bài 2. Dùng sklearn để xây dựng hàm tuyến tính, cũng như print ra giá trị của các biến số. 

x=np.array(datasource['Height'].values)
y=np.array(datasource['Weight'].values)

reg = LinearRegression().fit(x.reshape(-1,1), y)
w1=reg.coef_[0]
w0=reg.intercept_
print('w1 :', w1)
print('w0 :', w0)

y_pred = reg.intercept_ + reg.coef_[0]*x
r2Score = r2_score(y, y_pred)
print('R2 score: ', r2Score)
```

## Bài 4: Viết hàm đạo hàm.


### Hàm mất mát : $\frac{1}{N} \sum \limits _{i=1} ^N(y_{i} - \hat{y}_{i})^2 =\frac{1}{N} \sum \limits _{i=1} ^N(y_{i} - ( w_{1}*x_{i} +  w_{0}) )^2 $

### Đạo hàm theo $w_{1} : \frac{2}{N} \sum \limits _{i=1} ^N -x_{i}(y_{i} - ( w_{1}*x_{i} +  w_{0}) )$

### Đạo hàm theo $w_{0} : \frac{2}{N} \sum \limits _{i=1} ^N (y_{i} - ( w_{1}*x_{i} +  w_{0}) )$

```python
# Bài 4 Viết hàm đạo hàm.

# Tính giá trị đạo hàm hàm mất mát.

def D_w1function (x, y, w_1 ,w_0 ): 
    n = len(x)
    y_pred=w_1 * x + w_0
    D_m = (-2/n) * sum(x * (y - y_pred))
    return D_m
def D_w0function (x, y, w_1 ,w_0): 
    n = len(x)
    y_pred=w_1 * x + w_0
    D_m = (-2/n) * sum(y - y_pred)
    return D_m
```

```python
# Bài 5 : Xây dựng hàm gradient descent để tìm lời giải cho hàm tuyến tính, và print ra giá trị.

def gradient_descent(alpha,w_init,iteration,tolerance):
    w_list = [w_init]
    for it in range(iteration):
        w_0 = w_list[-1][0] - alpha * D_w0function(x,y,w_list[-1][1],w_list[-1][0])
        w_1 = w_list[-1][1] - alpha * D_w1function(x,y,w_list[-1][1],w_list[-1][0])
        if (w_0 - w_list[-1][0] < tolerance) or (w_1 - w_list[-1][1] < tolerance):
            break
        w_list.append([w_0,w_1])
    return w_list,it
```

```python
alpha = 0.00001
iteration = 1000
tolerance = 0.00001
w_init=[-400,-400]

w_list,it = gradient_descent(alpha,w_init, iteration,tolerance)
print('10 lần lặp cuối cùng có giá trị : ')
for i in range (-10,-1):
    print('Lần ',len(w_list)+i," : [",w_list[i][0],w_list[i][1],']')
    
print('\n\nGiá trị tối ưu nhất là :')
print('w1 :',w_list[-1][1])
print('w0 :',w_list[-1][0])
print( 'Số vòng lặp :',it)
```

```python
# Bài 6 :Vẽ các điểm và đường thẳng tuyến tính đã giải bằng gradient descent.

y_GradientDescent =  w_list[-1][1] * x+ w_list[-1][0]
plt.scatter(x, y, 50 ,colors)
plt.xlabel('Chiều cao (inch)')
plt.ylabel('Cân nặng (pound)')
plt.plot(x,y_GradientDescent, c = 'r', alpha=0.5)
```
