# Bài 4: Viết hàm đạo hàm.


### $\frac{1}{N} \sum \limits _{i=1} ^N(y_{i} - \hat{y}_{i})^2 =\frac{1}{N} \sum \limits _{i=1} ^N(y_{i} - ( w_{1}*x_{i} +  w_{0}) )^2 $

### Đạo hàm theo $w_{1} : \frac{2}{N} \sum \limits _{i=1} ^N -x_{i}(y_{i} - ( w_{1}*x_{i} +  w_{0}) )$

### Đạo hàm theo $w_{0} : \frac{2}{N} \sum \limits _{i=1} ^N (y_{i} - ( w_{1}*x_{i} +  w_{0}) )$

```python
# Bài 4 Viết hàm đạo hàm.
# Tính giá trị đạo hàm hàm mất mát

def D_w1function (x, y, w_1, w_0): 
    n = len(x)
    y_pred=w_1 * x + w_0
    D_m = (-2/n) * sum(x * (y - y_pred))
    return D_m
def D_w0function (x, y, w_1, w_0): 
    n = len(x)
    y_pred=w_1 * x + w_0
    D_m = (-2/n) * sum(y - y_pred)
    return D_m
```

```python
def update_w01(x,y,w_1,w_0,learning_rate):
    
    w_1_temp = D_w1function(x,y,w_1,w_0)
    w_0_temp = D_w0function(x,y,w_1,w_0)
        
    w_1 = w_1 - (w_1_temp*learning_rate)
    w_0 = w_0 - (w_0_temp*learning_rate)
    
    return w_1, w_0
```
