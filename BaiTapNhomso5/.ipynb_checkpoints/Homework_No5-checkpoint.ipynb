```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
#Cau1
df = pd.read_csv('weight-height.csv')
df.head()
```

```python
#Cau2
x = np.array(df['Height'].values)
y = np.array(df['Weight'].values)

reg = LinearRegression().fit(x.reshape(-1,1),y)
print('w1 using Sklearn is: ', reg.coef_[0])
print('w0 using Sklearn is: ', reg.intercept_)

y_pred = reg.intercept_ + reg.coef_[0]*x
r2Score = r2_score(y,y_pred)
print("R2 score is: ", r2Score)
```

```python
#Cau3
n = np.size(x)
colors = np.random.rand(n)
area = 50
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(x,y,area, colors, alpha = 0.5)
plt.plot(x,y_pred)
```

```python
#Cau4
def function_loss_derivative():
    n = len(x)
    w_1d = -(2/n)*sum(x*(y-y_pred))
    w_0d = -(2/n)*sum(y-y_pred)
    return w_1d, w_0d

function_loss_derivative()
```

```python
#Cau5
def gradientDescent():
    w1_new = w0_new = 0
    iter = 1000
    n = len(x)
    learning_rate = 0.001
    for i in range(iter):
        w_0d,w_1d = function_loss_derivative()
        w1_new = w1_new - learning_rate * w_1d
        w0_new = w0_new - learning_rate * w_0d
        print("w0 {}, w1 {}, iterations {}".format(w0_new,w1_new,i))
gradientDescent()
```
